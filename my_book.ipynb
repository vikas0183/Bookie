```python
df_books_average_rating = pd.read_json("front_end_data/Books average Rating.json")
df_books_average_rating.head()
```

```python
df_books_average_rating.sort_values("Average-Rating", ascending=False, inplace=True)
```

```python
df_books_average_rating.head()
```

```python
df_books = pd.read_csv("datasets/Books.csv")
df_books.head()
```

```python
df_merged_average_master = df_books_average_rating.merge(df_books, on='ISBN',how='inner')
df_merged_average_master.head()
```

```python
df_merged_average_master.columns
```

```python
df_merged_average_master.to_json("Master_average_data.json")
```
