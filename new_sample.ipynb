```python
import pandas as pd
```

```python
df_book_vectors = pd.read_json("Book_vectors.json")
df_book_vectors.head()
```

```python
df_master = pd.read_csv("datasets/Books.csv")
df_master.head()
```

```python
df_merged = pd.merge(df_book_vectors, df_master, on="Book-Title",how="inner")
df_merged.head()
```

```python
df_merged.to_json("Master_Book_Vectors.json")
```

```python
df_merged.drop_duplicates(subset="Book-Title", inplace=True)
```
